#### **Order analysis**

```python
zone_orders = df.groupby('Zone')["Orders"].sum().sort_values(ascending=False)
state_orders = df.groupby('State')["Orders"].sum().sort_values(ascending=False).head(5)
Product_category_orders = df.groupby('Product_Category')["Orders"].sum().sort_values(ascending=False).head()
Product_orders = df.groupby('Product_ID')["Orders"].sum().sort_values(ascending=False).head()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.pie(zone_orders, labels=zone_orders.index ,autopct='%1.1f%%', colors=sns.color_palette('Set1'))
plt.title('zone vs order')

plt.subplot(2,2,2)
plt.pie(state_orders, labels=state_orders.index ,autopct='%1.1f%%', colors=sns.color_palette('Set1'))
plt.title('State vs order')

plt.subplot(2,2,3)
plt.pie(Product_category_orders, labels=Product_category_orders.index ,autopct='%1.1f%%', colors=sns.color_palette('Set1'))
plt.title('Product_category vs order')

plt.subplot(2,2,4)
plt.pie(Product_orders, labels=Product_orders.index ,autopct='%1.1f%%', colors=sns.color_palette('Set1'))
plt.title('Product vs order')

plt.tight_layout()
plt.show()
```

### **Sales Analysis**

```python
fig, axarr = plt.subplots(4 , 2, figsize=(12,18))

df.groupby('State')['Amount'].sum().sort_values(ascending=False).plot.barh(ax=axarr[0][0], fontsize=10, color=sns.color_palette('Set1'))
axarr[0][0].set_title('location vs price',fontsize=10)

df.groupby('Zone')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=10, color=sns.color_palette('Set2'))
axarr[0][1].set_title('Zone vs price',fontsize=10)

df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).plot.barh(ax=axarr[1][0], fontsize=10, color=sns.color_palette('Set3'))
axarr[1][0].set_title('Product_Category vs price',fontsize=10)

df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10).plot.bar(ax=axarr[1][1], fontsize=10, color=sns.color_palette('Greens'))
axarr[1][1].set_title('Product_ID vs price',fontsize=10)

df.groupby('Orders')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[2][0], fontsize=10, color=sns.color_palette('Blues'))
axarr[2][0].set_title('Orders vs price',fontsize=10)

df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[2][1], fontsize=10, color=sns.color_palette('summer'))
axarr[2][1].set_title('Age Group vs price',fontsize=10)

df.groupby('Gender')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=10, color='indigo')
axarr[3][0].set_title('Gender vs orders',fontsize=10)

df.groupby('Marital_Status')['Amount'].sum().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=10, color='lightpink')
axarr[3][1].set_title('car_Marital_Status_age vs price',fontsize=10)
plt.subplots_adjust(hspace=0.8)
plt.subplots_adjust(wspace=0.5)
sns.despine()
plt.show()
```

## **Conclusion:**

- Regional Impact: Diwali sales are particularly strong in states like Uttar pradesh, Maharashtra, Karnataka and Delhi. This information can help in targeting marketing efforts more effectively.

- Zone-wise Strategy: Focus on promotional activities and inventory planning in the Central zone, as it leads in total sales during Diwali.

- Product Focus: Prioritize marketing and stock levels for products in the food, clothing and apparel, Electronic and Gadgests,  and Footwear and shoes, as they are the top performers.

- Product ID Insights: Products with IDs like P00265242, P00110942, P00184942, and P00112142 have consistently high demand, and efforts should be made to promote and maintain their availability.

- Order Frequency Strategy: Encourage customers to purchase more items in a single order by offering bundled deals or discounts for multiple items.

- Age Group Targeting: Target marketing campaigns towards customers in the Adult age group, as they contribute significantly to Diwali sales.

- Gender and Marital Status Targeting: Focus on single females, as they contribute the most to Diwali sales. Consider targeted promotions for different gender and marital status
